```python
import requests
import bs4
url = 'https://www.beatlesbible.com/songs/'
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text)

songs = [li.text.rsplit(' (')[0] for li in soup.find_all('li')[38:379]]
songs = ['The ' + s if s.endswith(', The') else s for s in songs ]

for song in songs:
    print(song)
```
